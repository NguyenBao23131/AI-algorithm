```python
!pip install rasa[full]
!pip install nest-asyncio==1.5.1

!rasa init --no-prompt

# Define the training data for the chatbot
training_data = """
version: "2.0"
nlu:
- intent: greet
  examples: |
    - hello
    - hi
    - hey there
- intent: good_bye
  examples: |
    - bye
    - goodbye
    - see you later
- intent: thanks
  examples: |
    - thanks
    - thank you
    - appreciate it
responses:
  utter_greet:
  - text: "Hello! How can I help you today?"
  utter_good_bye:
  - text: "Goodbye!"
  utter_default:
  - text: "I'm sorry, I didn't understand. Can you please rephrase?"
"""

# Write the training data to a file
with open("data/training_data.yml", "w") as f:
    f.write(training_data)

# Train the chatbot
!rasa train

# Start the Rasa server
!rasa run --enable-api --cors "*"
```
